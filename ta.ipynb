Q : Create a Student Performance Analyzer.
Expected behavior:
1. Read student marks from a CSV file (name, marks)
2. Store students using a class
3. Calculate average marks
4. Print student name and result:
○ Pass if marks ≥ 40
○ Fail otherwise

```python
import csv

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def result(self):
        if self.marks >= 40:
            print("Pass")
        else:
            print("Fail")

students = []

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)  # skip header

    for row in reader:
        name = row[0]
        marks = int(row[1])

        student = Student(name, marks)
        students.append(student)

for s in students:
    print(s.name)
    s.result()
    

```
