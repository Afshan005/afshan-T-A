Q : Create a Student Performance Analyzer.
Expected behavior:
1. Read student marks from a CSV file (name, marks)
2. Store students using a class
3. Calculate average marks
4. Print student name and result:
○ Pass if marks ≥ 40
○ Fail otherwise

```python
import csv

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def result(self):
        if self.marks >= 40:
            print("Pass")
        else:
            print("Fail")

students = []

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader)

    for row in reader:
        name = row[0]
        marks = int(row[1])

        student = Student(name, marks)
        students.append(student)

for s in students:
    print(s.name)
    s.result()
    

```

Q : Build a Simple Employee Payroll System. 
Expected behavior: 
1. Read employee data from a CSV file (name, salary) 
2. Store employees using a class 
3. Apply 10% tax deduction 
4. Print final salary

```python
import csv 
class Employee: 
def __init__(self, name, salary): 
self.name = name 
self.salary = salary 
def final_salary(self): 
tax = self.salary * 0.10 
return self.salary - tax 
 
employees = {} 
with open("employees.csv", "r") as file: 
    reader = csv.reader(file) 
    for row in reader: 
        name = row[0] 
        salary = row[1] 
        if name not in employees: 
            employees[name] = Employee(name, salary) 
for emp in employees: 
    print(emp.name) 
    print(emp.final_salary())
```
